```python
from qspectra import *
```

```python
electronic_fmo = np.array(np.mat("""
    12400 -87.7 5.5 -5.9 6.7 -13.7 -9.9;
    -87.7 12520 30.8 8.2 0.7 11.8 4.3;
    5.5 30.8 12200 -53.5 -2.2 -9.6 6.;
    -5.9 8.2 -53.5 12310 -70.7 -17. -63.3;
    6.7 0.7 -2.2 -70.7 12470 81.1 -1.3;
    -13.7 11.8 -9.6 -17. 81.1 12620 39.7;
    -9.9 4.3 6. -63.3 -1.3 39.7 12430
    """))

dipoles_fmo = np.array([d / np.linalg.norm(d) for d in
    np.array([[3.019, 3.442, 0.797, 3.213, 2.969, 0.547, 1.983],
              [2.284, -2.023, -3.871, 2.145, -2.642, 3.562, 2.837],
              [1.506, 0.431, 0.853, 1.112, -0.661, -1.851, 2.015]]).T])

fmo_hamiltonian = ElectronicHamiltonian(electronic_fmo, 0, 100,
                                        DebyeBath(CM_K * 77, 35, 106), dipoles_fmo)

dynamical_model = RedfieldModel(fmo_hamiltonian, hilbert_subspace='gef', unit_convert=CM_FS)
```

```python
fmo_hamiltonian.E('e')
```

```python
times = np.linspace(-1000, 1000, num=2000)
freq, pump1_fft = fourier_transform(times, pump1(times, 12500), rw_freq=12500,
                                    reverse_freq=True, unit_convert=CM_FS,
                                    positive_time_only=False)
freq, pump2_fft = fourier_transform(times, pump2(times, 12500), rw_freq=12500,
                                    reverse_freq=True, unit_convert=CM_FS,
                                    positive_time_only=False)
```

```python
plot(freq, np.abs(pump1_fft), freq, np.abs(pump2_fft))
for E in fmo_hamiltonian.E('e'):
    plot([E, E], [0, 0.01], 'k-')
xlim(12000, 13000)
```

```python
%%time
pump = GaussianPulse(12500, 100, scale=1e-3, freq_convert=CM_FS)
t, rho0 = simulate_two_pulses(dynamical_model, [pump, pump],
                             time_extra=1000, exact_isotropic_average=True,
                             rtol=1e-2)

pump1 = GaussianPulse(12200, 100, scale=1e-3, freq_convert=CM_FS)
pump2 = GaussianPulse(12600, 100, scale=1e-3, freq_convert=CM_FS)
t, rho = simulate_two_pulses(dynamical_model, [pump1, pump2],
                             time_extra=1000, exact_isotropic_average=True,
                             rtol=1e-2)
```

```python
title('Laser excitation')
plot(t, rho0.reshape(-1, 8, 8)[:, 1, 2].real, 'k-', linewidth=1)
plot(t, rho.reshape(-1, 8, 8)[:, 1, 2].real, 'b-', linewidth=1)
xlabel('Time (fs)')
ylabel('Population')
xlim(pump1.t_init, 1000)
legend(np.concatenate([['Laser'], ['Site {}'.format(n) for n in xrange(1, 8)]]))
```

```python
T = t[t > pump1.t_final][:100]
rho_after = rho[t > pump1.t_final][:100]
rho0_after = rho0[t > pump1.t_final][:100]
```

```python
def probe(rho_after):
    X = []
    for cur_rho in rho_after:
        f, Xi = impulsive_probe(dynamical_model, cur_rho, 5000, 'xx', exact_isotropic_average=True)
        X.append(Xi)
    X = np.array(X)
    return f, X
```

```python
%%time
f, X0 = probe(rho0_after)
```

```python
%%time
f, X = probe(rho_after)
```

```python
for Xi in X:
    plot(f, Xi.real)
xlim(12000, 12800)
```
