```python
from qspectra import *
```

```python
ham = ElectronicHamiltonian(np.array([[12881, 120], [120, 12719]]),
                            bath=DebyeBath(CM_K * 77, 35, 106),
                            dipoles=[[1, 0, 0], [2 * np.cos(.3), 2 * np.sin(.3), 0]])
dimer = RedfieldModel(ham, hilbert_subspace='gef', discard_imag_corr=True, unit_convert=CM_FS)
```

```python
dimer.time_step
```

Parameters (other than bath) from Yuen-Zhou et al., PNAs 108, (2011)

## 3rd order response in time domain

```python
%%time
(t1, t2, t3), X = third_order_response(dimer, coherence_time_max=1000,
                                       population_times=np.linspace(0, 500, 25),
                                       geometry='-++', polarization='xxxx')
```

```python
contourf(t1, t3, X[:,0,:].real, 20)
```

```python
contourf(t1, t3, X[:,-1,:].real, 20)
```

## Convert to 2D spectra

```python
f1, X_ftt = fourier_transform(t1, X, 0, rw_freq=dimer.rw_freq, unit_convert=CM_FS)
f3, X_ftf = fourier_transform(t3, X_ftt, 2, rw_freq=dimer.rw_freq, unit_convert=CM_FS)
```

```python
contourf(f1, f3, X_ftf[:,0,:].real, 30)
xlim(12300, 13300)
ylim(12300, 13300)
```

```python
contourf(f1, f3, X_ftf[:,-1,:].real, 30)
xlim(12300, 13300)
ylim(12300, 13300)
```

Identify/plot cross-peaks

```python
alpha, beta = np.argmin(np.abs(f1 - ham.E('e').reshape(-1, 1)), 1)
```

```python
plot(t2, X_ftf[alpha, :, beta].real, '-s')
plot(t2, X_ftf[beta, :, alpha].real, '-s')
```
